### 通过 sum 进行降维操作

```python
X = torch.arange(24).reshape(2, 3, 4)
Y = X.clone()
# 将张量乘以或加上一个标量不会改变张量的形状，其中张量的每个元素都将与标量相加或相乘。

# 通过 sum 进行降维操作
A_sum_axis0 = X.sum(axis=0) # shape 的 轴0 消失
A_sum_axis1 = X.sum(axis=1)# shape 的 轴1 消失
A_sum_axis2 = X.sum(axis=2)# shape 的 轴2 消失
# X, A_sum_axis0, A_sum_axis0.shape, A_sum_axis1, A_sum_axis1.shape, A_sum_axis2, A_sum_axis2.shape
#X, Y, X*Y, X, X+2, X.sum()-(X+2).sum()
# X, X.sum(axis=[0,1])

# 非降维求和 keepdim=True
A_sum_axis00 = X.sum(axis=0, keepdim=True) # shape 的 轴0 消失
A_sum_axis0.shape, A_sum_axis00.shape
```

## 2.3.9. 矩阵-矩阵乘法

```python
B=torch.arange(12,dtype=torch.float32).reshape(4,3)
v = torch.tensor([0,1,2,3], dtype=torch.float32)
vv = torch.tensor([0,1,2,3], dtype=torch.float32)+4
vvv = torch.tensor([0,1,2,3], dtype=torch.float32)+4*2
C = torch.mm(A,B)  # mm = metric * metric
# 矩阵乘法 = 后者矩阵退化成向量的向量, 可以看成 矩阵-向量积(向量是后者列轴维度)
A, B, torch.mv(A,v), torch.mv(A,vv), torch.mv(A,vvv), C
```
